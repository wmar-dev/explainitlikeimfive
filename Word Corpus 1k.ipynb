```python
import re
from collections import Counter
import nltk
from nltk.corpus import reuters
from nltk.corpus import brown
import random

%load_ext autoreload
%autoreload 2   
```

```python
nltk.download('reuters')
nltk.download('brown')
```

```python
letters_regex = re.compile(r'^[A-Za-z]+$')
```

```python
%%time
word_count = Counter((word.lower() for cat in reuters.categories() for word in reuters.words(categories=cat) if letters_regex.search(word)))
reuters_top1k_words = [w[0] for w in word_count.most_common()[:1000]]
reuters_top1k_words
```

```python
word_count = Counter((word.lower() for cat in brown.categories() for word in brown.words(categories=cat) if letters_regex.search(word)))
brown_top1k_words = [w[0] for w in word_count.most_common()[:1000]]
```

```python
diff = set(brown_top1k_words).difference(reuters_top1k_words)
print(len(diff))
print(random.sample(list(diff), 10))
print(diff)
```

```python
diff = set(reuters_top1k_words).difference(brown_top1k_words)
print(len(diff))
print(random.sample(list(diff), 10))
print(diff)
```
